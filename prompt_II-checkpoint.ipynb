### Modeling


With your (almost?) final dataset in hand, it is now time to build some models.  Here, you should build a number of different regression models with the price as the target.  In building your models, you should explore different parameters and be sure to cross-validate your findings.

```python
lr_1_feature  = LinearRegression(fit_intercept=False).fit(CarSales[['year']], CarSales['price'])
lr_2_features = LinearRegression(fit_intercept=False).fit(CarSales[['year','odometer']], CarSales['price'])
lr_3_features = LinearRegression(fit_intercept=False).fit(CarSales[['year','odometer','cylinders']], CarSales['price'])
lr_4_features = LinearRegression(fit_intercept=False).fit(CarSales[['year','odometer','cylinders','fuel']], CarSales['price'])
lr_5_features = LinearRegression(fit_intercept=False).fit(CarSales[['year','odometer','cylinders','fuel','title_status']], CarSales['price'])
```

### Evaluation

With some modeling accomplished, we aim to reflect on what we identify as a high quality model and what we are able to learn from this.  We should review our business objective and explore how well we can provide meaningful insight on drivers of used car prices.  Your goal now is to distill your findings and determine whether the earlier phases need revisitation and adjustment or if you have information of value to bring back to your client.

```python
pred1 = lr_1_feature.predict(CarSales[['year']])
pred2 = lr_2_features.predict(CarSales[['year','odometer']])
pred3 = lr_3_features.predict(CarSales[['year','odometer','cylinders']])
pred4 = lr_4_features.predict(CarSales[['year','odometer','cylinders','fuel']])
pred5 = lr_5_features.predict(CarSales[['year','odometer','cylinders','fuel','title_status']])



error_dict = {'Features': ['1 Feature', '2 Features', '3 Features', '4 Features', '5 Features'],
'MSE': [mean_squared_error(CarSales['price'], i) for i in [pred1, pred2, pred3,pred4,pred5]],
'MAE': [mean_absolute_error(CarSales['price'], i) for i in [pred1, pred2, pred3,pred4,pred5]]}

error_df = pd.DataFrame(error_dict).set_index('Features')



error_df
```
